```python
num_cols = df.select_dtypes(include="number").columns
cat_cols = df.select_dtypes(exclude="number").columns

len(num_cols), len(cat_cols)
```

```python
for col in num_cols:
    df[col].hist(bins=30)
    plt.title(col)
    plt.show()
```

```python
for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), cmap="coolwarm")
plt.show()
```

```python
df[num_cols].corr()["SalePrice"].sort_values(ascending=False).head(10)
```

```python
top = df[num_cols].corr()["SalePrice"].abs().sort_values(ascending=False).index[1:6]

for col in top:
    sns.scatterplot(x=df[col], y=df["SalePrice"])
    plt.title(col)
    plt.show()
```

```python
for col in cat_cols[:5]:
    sns.boxplot(x=df[col], y=df["SalePrice"])
    plt.xticks(rotation=45)
    plt.show()
```

##Conclusion

Dataset size: The dataset contains 1460 rows and 81 columns.

Missing values: Several columns contain missing values, with the highest missing counts observed in Alley, PoolQC, Fence, MiscFeature, and LotFrontage.

Strongest correlations with SalePrice: SalePrice shows strong positive correlation with OverallQual, GrLivArea, GarageCars, GarageArea, and TotalBsmtSF. These features represent house quality and living area, which have the greatest impact on price.

Outliers: Boxplots reveal noticeable outliers in several numerical features such as GrLivArea, LotArea, GarageArea, and porch-related variables. Many porch features contain mostly zero values with a few extreme values, indicating right-skewed distributions.
